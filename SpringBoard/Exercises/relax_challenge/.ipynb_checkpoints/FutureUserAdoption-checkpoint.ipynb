Defining an "adopted user" as a user who has logged into the product on three separate days in at least one seven day period, identify which factors predict future user adoption.

It would be helpful to join the two tables on user ID.

```python
# Joining two dataframes

df1 = df1.rename(columns={"object_id": "user_id"})
df = pd.merge(df1, df2, how='outer', on='user_id')
```

```python
df
```

The strategy:
    
The goal is to predict user adoption, which is defined in the problem. I will approach this by using binary classification models, and the two classes will be "not adopted user" and "adopted user". First I will need to use the data given to determine which of the two classes each of the users in the data falls under. Once they are all labeled, we can begin to use the data to build models that are able to predict future user adoption. 

An adopted user is a user who's logged into the product on several days in at least one seven-day period. To correctly identify each user as an adopted user or not, we'll have to iterate through each user and examine the timestamps of all of their logins. Based on these timestamps, we will be able to tell if each user meets the criteria of an adopted user.

For each user, retrieve all of their login time stamps, and iterate through each 3 day grouping. If the there is less than a seven day difference between the first and third day of each three-day grouping, then the user can be said to be an adopted user. Dates come sorted already so no need for us to do that. If user has less than 2 total logins then they can be immediately be ruled as not an adopted user. 

```python
import datetime

def isAdopted(user_id):
    timestamps = df['time_stamp'][df['user_id'] == user_id]
    if len(timestamps) <= 2:
        return 0
    
    dates = [datetime.datetime(int(timestamp[:4]), int(timestamp[5:7]), int(timestamp[8:10])) for timestamp in timestamps]
    for i in range(len(dates) - 2):
        daysApart = dates[i+2] - dates[i]
        if daysApart.days <= 7:
            return 1

    return 0 
```

```python
adoptedUsers = []

for i in range(12000):
    adoptedUsers.append(isAdopted(i+1))
```

```python
df1['Adopted'] = adoptedUsers
df1
```

# Feature Engineering

creation_time looks to be a promising feature but its str format does not make it readily available to work with. I instead would rather have a feature that tells us how much time has elapsed between "now" and when the user created an account. 

```python
daysOld = []

# Arbitrary "now" that dates somewhat recently to the most recent date in the data set.
now = datetime.datetime(2015, 1, 1)

for i in range(len(df1)):
    timestamp = df1['creation_time'][i]
    creationtime = datetime.datetime(int(timestamp[:4]), int(timestamp[5:7]), int(timestamp[8:10]))
    daysApart = now - creationtime
    daysOld.append(daysApart.days)
```

```python
df1['daysOld'] = daysOld
```

'name' and 'email' should not be great predictive features at all because each user should have a name and email that is unique to them. Even if it's possible for two people to have the same first and last name, there's no way that having a certain name will cause users to be more likely to be "adopted users". 

creation_source can be encoded into dummy columns because there are only 5 unique values of it.

last_session_creation_time appears to have quite an amount of missing values. If there is a missing value in this column, it simply means that the user has not logged in a single time since creating their account. Let's instead engineer this feature to be "days since last login". For users who haven't logged in since creating their account, this value will simply be their 'daysOld' value.

```python
daysLastLogin = []

for i in range(len(df1)):
    timestamp = df1['last_session_creation_time'][i]
    if np.isnan(timestamp):
        daysLastLogin.append(daysOld[i])
    else:
        date = datetime.datetime.utcfromtimestamp(timestamp)
        daysApart = now - date
        daysLastLogin.append(daysApart.days)
```

```python
df1['daysLastLogin'] = daysLastLogin
```

The next two features are fine as they are, now we come across 'org_id'. Let's see how many unique values of this column there are.

```python
df1['org_id'].value_counts()
```

While org_id is a numeric value, this is in reality a categorical feature. It would be misleading to leave the column as is, as there is no such thing as an org being "greater" than another org. Dummy encoding would result in too many new columns so I would process this feature with binary encoding.

```python
import category_encoders as ce

encoder = ce.BinaryEncoder(cols=['org_id'],return_df=True)

org_encoded = encoder.fit_transform(df1) 
org_encoded = org_encoded[['org_id_0', 'org_id_1', 'org_id_2', 'org_id_3', 'org_id_4', 'org_id_5',
                           'org_id_6', 'org_id_7', 'org_id_8', 'org_id_9']]
```

'invited_by_user_id' also has a number of missing values. This makes sense, as not every user will have necessarily been invited to start using the service. Let's first see how many unique values of this column there are.

```python
df1['invited_by_user_id'].value_counts()
```

If we were to encode this column into dummy columns, 2500 additional columns would be generated. Were I to have the computing power to handle this I'd go for it, but I don't so I'm going to simplify this feature. I'm simply going engineer a feature called "invited", which will tell us if the user was invited or not.

```python
invited = []

for i in range(len(df1)):
    if np.isnan(df1['invited_by_user_id'][i]):
        invited.append(0)
    else:
        invited.append(1)
```

```python
df1['invited'] = invited
```
